Remember, the "object" data type is usually for strings.
Only four columns were successfully converted to a data type—revenue, runtime, vote_average, and
vote_count—but taking a look through the DataFrame, we can see some columns that should be specific data types.

We'll just go down the list and convert the data types for each of the six columns that need to be converted.

Before we convert the "adult" and "video" columns, we want to check that all the values are either True or False.

```python
kaggle_metadata['adult'].value_counts()
```

# Remove Bad Data


```python
kaggle_metadata[~kaggle_metadata['adult'].isin(['True','False'])]
```

In fact, since we probably don't want to include adult movies in the hackathon dataset, 
we'll only keep rows where adult is False, and then drop the "adult" column.



```python
kaggle_metadata = kaggle_metadata[kaggle_metadata['adult'] == 'False'].drop('adult',axis='columns')
```

```python
kaggle_metadata["video"].value_counts()
```

# Convert Data Types

```python
kaggle_metadata['video'] == 'True'
```

The above code creates the Boolean column we want. We just need to assign it back to video:

```python
kaggle_metadata['video'] = kaggle_metadata['video'] == 'True'
```

For the numeric columns, we can just use the to_numeric() method from Pandas. 
We'll make sure the errors= argument is set to 'raise', so we'll know if there's 
any data that can't be converted to numbers.

```python
kaggle_metadata['budget'] = kaggle_metadata['budget'].astype(int)
kaggle_metadata['id'] = pd.to_numeric(kaggle_metadata['id'], errors='raise')
kaggle_metadata['popularity'] = pd.to_numeric(kaggle_metadata['popularity'], errors='raise')
```

Finally, we need to convert release_date to datetime. Luckily, Pandas has a built-in function for that as well: 
    to_datetime().

```python
kaggle_metadata['release_date'] = pd.to_datetime(kaggle_metadata['release_date'])
```

# Reasonability Checks on Ratings Data

Lastly, we'll take a look at the ratings data. We'll use the info() method on the DataFrame. Since the ratings dataset has so many rows, we need to set the null_counts option to True.

```python
ratings.info(null_counts=True)
```

We'll specify in to_datetime() that the origin is 'unix' and the time unit is seconds.

```python
pd.to_datetime(ratings['timestamp'], unit='s')
```
